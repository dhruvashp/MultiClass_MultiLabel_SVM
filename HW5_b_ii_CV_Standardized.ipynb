For standardized features, the portion of range selection technically should be repeated, however we assume that with standardization the range of C and gamma will approximately be obtained similar to the ones obtained for the raw data. Again this is not an ideal assumption, but to save run times, we assume this nonetheless

Thus the C and gamma value range for standardized features will be taken same as the range taken for raw features.

Additionally, Standardization will be done on the entire dataset (train + test) only ONCE, like it has been done in previous homeworks.

Thus Standardization will be performed once, on the entire dataset as a whole (train + test), and then that dataset will be again split into train and test, train used for CV and test used to obtain the actual test error

SO, only ONCE standardization, over ENTIRE dataset, and optimizing over SAME C and gamma values as before, will be done here

NOTE : I have assumed that just like the entire dataset has been given, normalized to us, standardizing here means that we are simply standardizing the ENTIRE dataset once and treating it as a 'standardized version' of the original dataset

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC

df_train = pd.read_csv('Train_Data.csv',index_col=0)
df_test = pd.read_csv('Test_Data.csv',index_col=0)
print(df_train)
print(df_test)
```

```python
features_train = df_train.drop(columns=['Family','Genus','Species'])
features_test = df_test.drop(columns=['Family','Genus','Species']) 

family_train = pd.DataFrame(df_train['Family'])
genus_train = pd.DataFrame(df_train['Genus'])
species_train = pd.DataFrame(df_train['Species'])

family_test = pd.DataFrame(df_test['Family'])
genus_test = pd.DataFrame(df_test['Genus'])
species_test = pd.DataFrame(df_test['Species'])
```

```python
print(features_train)
print(features_test)
```

Standardization
We'll standardize only over the features, obviously, once, in entirety

```python
whole_features = pd.concat([features_train,features_test])
whole_features.reset_index(drop=True,inplace=True)
print(whole_features)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(whole_features)

whole_features_scaled = scaler.transform(whole_features)

print(whole_features_scaled)
```

```python
whole_features_scaled_df = pd.DataFrame(whole_features_scaled,columns=whole_features.columns)
print(whole_features_scaled_df)
```

```python
features_train = whole_features_scaled_df.iloc[0:features_train.shape[0],:]  # Updating features_train with standardized values
features_test = whole_features_scaled_df.iloc[features_train.shape[0]:,:] # Updating features_test with standardized values
features_train.reset_index(drop=True,inplace=True)
features_test.reset_index(drop=True,inplace=True)
print(features_train)
print(features_test)
```

```python
print(family_train)
print(genus_train)
print(species_train)
```

```python
print(family_test)
print(genus_test)
print(species_test)
```

Cross Validation :
C values = {10^-2, 10^-1, 10^0, 10^1, 10^2, 10^3, 10^4, 10^5, 10^6, 10^7}
gamma values = {0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.4, 1.7, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 9.5, 10}

Jump in gamma values to save computational time

Performance will again be evaluated on the basis of Hamming Loss (can be also done on Exact Matching, here Hamming loss is selected, as such both can also be selected and somehow 'polled', again here only Hamming Loss is considered)

```python
from sklearn.model_selection import KFold
import warnings
warnings.filterwarnings('ignore')
kf = KFold(n_splits=10,shuffle=True)  # 10-fold CV
C = np.array([10**(-2),10**(-1),10**(0),10**1,10**2,10**3,10**4,10**5,10**6,10**7])
gamma = np.array([0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.4, 1.7, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 9.5, 10])

CV_errors = pd.DataFrame(index=C,columns=gamma)

for i in np.arange(0,C.size):
    
    print(i)   # For status check, run time too long
    
    for j in np.arange(0,gamma.size):
        
        cv_ite = -1
        
        cv_loc_err = np.zeros(10)
        
        for train, test in kf.split(features_train):
            cv_ite = cv_ite + 1
            X_train,X_test = features_train.iloc[train,:],features_train.iloc[test,:]
            yf_train,yf_test = family_train.iloc[train,:],family_train.iloc[test,:]
            yg_train,yg_test = genus_train.iloc[train,:],genus_train.iloc[test,:]
            ys_train,ys_test = species_train.iloc[train,:],species_train.iloc[test,:]
            
            clf_f = SVC(C=C[i],gamma=gamma[j],kernel='rbf',decision_function_shape='ovr').fit(X_train,yf_train)
            clf_g = SVC(C=C[i],gamma=gamma[j],kernel='rbf',decision_function_shape='ovr').fit(X_train,yg_train)
            clf_s = SVC(C=C[i],gamma=gamma[j],kernel='rbf',decision_function_shape='ovr').fit(X_train,ys_train)
            
            yf_pred_test = clf_f.predict(X_test)
            yg_pred_test = clf_g.predict(X_test)
            ys_pred_test = clf_s.predict(X_test)
            
            mis = 0
            
            for k in np.arange(0,X_test.shape[0]):
                if yf_pred_test[k] != yf_test.iloc[k,0]:
                    mis = mis + 1
                    
                if yg_pred_test[k] != yg_test.iloc[k,0]:
                    mis = mis + 1
                    
                if ys_pred_test[k] != ys_test.iloc[k,0]:
                    mis = mis + 1
            
            
            mis_percentage = (mis/(3*X_test.shape[0]))*100
            
            
            cv_loc_err[cv_ite] = mis_percentage
        
        
        
        CV_errors.iloc[i,j] = np.mean(cv_loc_err)

                
        
print('The CV test error estimates, averaged over 10 fold CV, for different C and gamma, C in rows and gamma in columns, is obtained thus : \n', CV_errors)
```

```python
CV_errors.to_csv('CV_errors_Matrix_Standardized.csv')
```

```python
minim = CV_errors.iloc[0,0]
for i in np.arange(0,CV_errors.shape[0]):
    for j in np.arange(0,CV_errors.shape[1]):
        if CV_errors.iloc[i,j] < minim:
            minim = CV_errors.iloc[i,j]
            min_C = C[i]
            min_gamma = gamma[j]

print('The minimum error obtained is : \n',minim,'%')
print('The C and gamma that minimizes this test error estimate over the selected C and gamma range is : \n',min_C,'and',min_gamma,'respectively')
```

```python
clf_fam = SVC(C=min_C,gamma=min_gamma,kernel='rbf',decision_function_shape='ovr').fit(features_train,family_train)
clf_gen = SVC(C=min_C,gamma=min_gamma,kernel='rbf',decision_function_shape='ovr').fit(features_train,genus_train)
clf_spe = SVC(C=min_C,gamma=min_gamma,kernel='rbf',decision_function_shape='ovr').fit(features_train,species_train)


y_fam_pre = clf_fam.predict(features_test)
y_gen_pre = clf_gen.predict(features_test)
y_spe_pre = clf_spe.predict(features_test)

mis = 0
for i in np.arange(0,features_test.shape[0]):
    if y_fam_pre[i] != family_test.iloc[i,0]:
        mis = mis + 1
    if y_gen_pre[i] != genus_test.iloc[i,0]:
        mis = mis + 1
    if y_spe_pre[i] != species_test.iloc[i,0]:
        mis = mis + 1
    
test_error = (mis/(3*features_test.shape[0]))*100
print('The actual test error obtained with the optimized C and gamma is : \n',test_error,'%')
```
