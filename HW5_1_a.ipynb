```python
import pandas as pd
import numpy as np
```

```python
raw_df = pd.read_csv('Frogs_MFCCs.csv')
print(raw_df)
```

```python
df = raw_df.drop(columns=['RecordID'])
print(df)
```

Changing string classes to numeric before splitting

```python
print(df.shape[0])
```

```python
for i in np.arange(0,df.shape[0]):
    if df.iloc[i,22] == 'Bufonidae':
        df.iloc[i,22] = 0
    elif df.iloc[i,22] == 'Dendrobatidae':
        df.iloc[i,22] = 1
    elif df.iloc[i,22] == 'Hylidae':
        df.iloc[i,22] = 2
    else:
        df.iloc[i,22] = 3


for i in np.arange(0,df.shape[0]):
    if df.iloc[i,23] == 'Adenomera':
        df.iloc[i,23] = 0
    elif df.iloc[i,23] == 'Ameerega':
        df.iloc[i,23] = 1
    elif df.iloc[i,23] == 'Dendropsophus':
        df.iloc[i,23] = 2
    elif df.iloc[i,23] == 'Hypsiboas':
        df.iloc[i,23] = 3        
    elif df.iloc[i,23] == 'Leptodactylus':
        df.iloc[i,23] = 4
    elif df.iloc[i,23] == 'Osteocephalus':
        df.iloc[i,23] = 5
    elif df.iloc[i,23] == 'Rhinella':
        df.iloc[i,23] = 6
    else:
        df.iloc[i,23] = 7
        

for i in np.arange(0,df.shape[0]):
    if df.iloc[i,24] == 'AdenomeraAndre':
        df.iloc[i,24] = 0
    elif df.iloc[i,24] == 'AdenomeraHylaedactylus':
        df.iloc[i,24] = 1
    elif df.iloc[i,24] == 'Ameeregatrivittata':
        df.iloc[i,24] = 2
    elif df.iloc[i,24] == 'HylaMinuta':
        df.iloc[i,24] = 3        
    elif df.iloc[i,24] == 'HypsiboasCinerascens':
        df.iloc[i,24] = 4
    elif df.iloc[i,24] == 'HypsiboasCordobae':
        df.iloc[i,24] = 5
    elif df.iloc[i,24] == 'LeptodactylusFuscus':
        df.iloc[i,24] = 6
    elif df.iloc[i,24] == 'OsteocephalusOophagus':
        df.iloc[i,24] = 7
    elif df.iloc[i,24] == 'Rhinellagranulosa':
        df.iloc[i,24] = 8
    else:
        df.iloc[i,24] = 9
        
        
print(df)
```

Thus the string classes have been converted to numeric, with the coding table seen above

Splitting into Train and Test

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, train_size = 0.7, shuffle = True) # Ideally stratification over each label must be done, this hasn't been done here as multi-label stratification may require us to write those codes from scratch

df_train.reset_index(drop=True, inplace = True)
df_test.reset_index(drop=True, inplace = True)

print(df_train)
print(df_test)
```

```python
df_train.to_csv('Train_Data.csv')
df_test.to_csv('Test_Data.csv')
```
