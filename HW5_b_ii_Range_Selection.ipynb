Selecting Range of C and gamma (Regularization Parameter and Kernel Co-efficient, respectively, as defined in scikit learn)

The Approach :

Before cross-validating over C and gamma, we need to pick ranges of C and gamma, appropriately, to cross-validate over.
As explained in the HW footnote, we'll do this as follows :

We will pick small values and large values (extremely small and extremely large) for C and gamma
Once, arbitrarily, these values have been picked, we will fit the data, the entire training data, over all the combinations of C and gamma

Once the data has been fit, we have two options : Calculating Training Errors via Hamming Loss or via Exact Match Loss
As such either, or even both could be picked. To make the problem simple, we'll pick the Hamming Loss.

Thus we'll calculate the hamming errors for each combination. We'll save the errors in a .csv and select then the appropriate range for C and gamma

Also, once the results have been obtained, we may write a piece of code that scans through the .csv file with errors and picks the end points for C and gamma

We'll keep the number of C and gamma small
NOTE : This isn't CV. CV will be done later.

C Values = {10^-5,10^-4,10^-3,10^-2,10^2,10^3,10^4,10^5,10^6,10^7}
gamma values = {0.0001,0.001,0.01,0.1,1,2,5,10,100,1000}

These 10 very small and very large values were randomly picked as such, with some footnote reference.

```python
C = np.array([10**(-5),10**(-4),10**(-3),10**(-2),10**(2),10**(3),10**(4),10**(5),10**(6),10**(7)])
gamma = np.array([0.0001,0.001,0.01,0.1,1,2,5,10,100,1000])
Range_Summary = pd.DataFrame(index = C,columns = gamma)

for i in np.arange(0,C.size):
    for j in np.arange(0,gamma.size):
        clf_fam = SVC(C=C[i],gamma=gamma[j],kernel='rbf',decision_function_shape='ovr').fit(features_train,family_train)
        clf_gen = SVC(C=C[i],gamma=gamma[j],kernel='rbf',decision_function_shape='ovr').fit(features_train,genus_train)
        clf_spe = SVC(C=C[i],gamma=gamma[j],kernel='rbf',decision_function_shape='ovr').fit(features_train,species_train)
        
        y_fam_pre = clf_fam.predict(features_train)
        y_gen_pre = clf_gen.predict(features_train)
        y_spe_pre = clf_spe.predict(features_train)
        
        mis = 0
        for k in np.arange(0,features_train.shape[0]):
            if y_fam_pre[k] != family_train.iloc[k,0]:
                mis = mis + 1
                
            if y_gen_pre[k] != genus_train.iloc[k,0]:
                mis = mis + 1
            
            if y_spe_pre[k] != species_train.iloc[k,0]:
                mis = mis + 1
            
        error_percentage = (mis/(3*features_train.shape[0]))*100
        
        Range_Summary.iloc[i,j] = error_percentage
        
        
print(Range_Summary)
Range_Summary.to_csv('Range_Selection.csv')
```

The errors have been obtained in percentage. Considering accuracy cut-off of 70%, we get an error cut-off of 30%

Thus range of C and gamma will be selected using minimum and maximum values such that in those ranges we get an error <= 30%

As such we can write a code for this, but it is not really needed as we can directly observe these values in the .csv extracted file and select the appropriate region hands-on

Observing the .csv file, Approximate max and min to CV over for C and gamma is obtained thus :

C min = 0.01
C max = 10^7

gamma min = 0.1
gamma max = 10


NOTE :
C min should be more than 0.01 based on .csv, however as for the 'min' space of C values 0.01 was the last value, it thus has been selected. It only gives error >= 30% for gamma = 0.1, for all the rest gamma's it is well under 30%

C max possibly could be larger than 10^7, but again it was the last upper bound entry, and thus was selected as the final range culminating boundary

gamma min could be possibly selected lower, however again issues with corresponding C range (the lower one) appear. Thus 0.1 was selected

gamma max could surely be taken till 1000, however according to the footnote, CV is done on gamma with gamma values changing linearly, and thus selecting a very large gamma range would increase run time to a large degree


Based on this analysis, for the subsequent CV, we will take values of C min, C max as 0.01 and 10^7 (logarithmic scale)
gamma min, gamma max as 0.1 and 10, linear scale
